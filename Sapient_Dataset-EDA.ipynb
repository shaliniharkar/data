```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
data= pd.read_csv('C:\\Users\\win8.1\\Desktop\\Case_Onsite_Modeling_Wine\\wine_dataset.csv')
data.head(10)
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(data['quality'],data['fixed_acidity'])
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.scatter(data['fixed_acidity'],data['pH'])
plt.xlabel('Acidity').set_size(20)
plt.ylabel('pH').set_size(20)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.pointplot(data['quality'],data['pH'], color='green')
plt.xlabel('Quality').set_size(20)
plt.ylabel('pH').set_size(20)
plt.show()
```

```python
plt.figure(figsize=(5,10))
sns.countplot(data['citric_acid'] > 0)
plt.xlabel('Citric Acid Content').set_size(20)
plt.ylabel('Count').set_size(20)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.barplot(data['quality'],data['total_sulfur_dioxide'])
plt.xlabel('Quality').set_size(20)
plt.ylabel('Total Sulfur Dioxide').set_size(20)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.pointplot(data['pH'].round(1),data['residual_sugar'], color='green')
plt.xlabel('pH').set_size(15)
plt.ylabel('Residual Sugar').set_size(15)
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.countplot(x=data['chlorides'].round(2))
plt.xlabel('Chlorides').set_size(15)
plt.ylabel('Count').set_size(15)
plt.show()
```

```python
sns.set()
pd.DataFrame.hist(data, figsize = [15,15], color='red')
plt.show()
```

```python
colormap = plt.cm.inferno
plt.figure(figsize=(12,12))
plt.title('Correlation of Features', y=1.05, size=15)
sns.heatmap(data.corr(),linewidths=0.1,vmax=1.0, square=True, cmap=colormap, linecolor='blue', annot=True)
plt.show()
```
